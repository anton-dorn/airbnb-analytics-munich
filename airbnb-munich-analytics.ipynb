# Step 2. Construct a Model

```python
# Run linear regression to find candidate variables for the dependent variable 'estimated revenue'
# based on their p-values with respect to the dependent variable

import statsmodels.api as sm
import numpy as np    # Numerical operations and NaN handling


# Define the dependent variable
dependent_var = 'estimated_revenue_l365d'

# Extract all numerical columns from the dataset
numeric_cols = df_filtered_occupancy.select_dtypes(include=['number']).columns.tolist()

# Remove the dependent variable from the list of independent variables
if dependent_var in numeric_cols:
    numeric_cols.remove(dependent_var)


print(f"P-values for each numerical independent variable with {dependent_var}:")
print("------------------------------------------------------------------")

results_list = []  # Store p-values for each independent variable

# Loop through each numerical independent variable
for col in numeric_cols:
    # Select the current independent variable and the dependent variable
    # and drop rows with missing values
    temp_df = df_filtered_occupancy[[col, dependent_var]].dropna()

    # Ensure sufficient data for regression
    if not temp_df.empty and len(temp_df) > 1:
        X_single = temp_df[[col]]          # Independent variable
        y_single = temp_df[dependent_var]  # Dependent variable

        # Add intercept term
        X_single = sm.add_constant(X_single)

        try:
            # Fit univariate OLS regression model
            model = sm.OLS(y_single, X_single)
            results = model.fit()

            # Extract p-value of the independent variable (excluding the intercept)
            if len(results.pvalues) > 1:
                p_value = results.pvalues.iloc[1]
                results_list.append({'Variable': col, 'P-value': p_value})
            else:
                results_list.append({'Variable': col, 'P-value': np.nan})
        except Exception:
            # Assign NaN if regression fails
            results_list.append({'Variable': col, 'P-value': np.nan})
    else:
        # Assign NaN if insufficient data
        results_list.append({'Variable': col, 'P-value': np.nan})

# Convert results to a DataFrame for clear presentation
p_values_df = pd.DataFrame(results_list)

# Sort variables by p-value (ascending), placing NaNs at the end
p_values_df = p_values_df.sort_values(by='P-value', na_position='last').reset_index(drop=True)

# Display the p-value table
display(p_values_df)

# We calculated the p-values of univariate regressions
# between each numerical independent variable and the dependent variable, estimated revenue,
# as an initial step in model construction.
# Several variables were found to be statistically significant at the 5% significance level (p < 0.05).
```

```python
# Build and estimate an initial multivariate OLS regression model to explain estimated revenue using selected numerical variables

# Updated list of independent variables after excluding insignificant ones
independent_vars = [
    'beds', 'number_of_reviews_ltm', 'accommodates', 'bedrooms',
    'number_of_reviews_ly', 'number_of_reviews_l30d', 'bathrooms',
    'maximum_nights', 'availability_365', 'maximum_maximum_nights',
    'minimum_maximum_nights', 'review_scores_cleanliness',
    'maximum_nights_avg_ntm', 'availability_30'
]


# Check whether the dependent variable and all independent variables exist in the DataFrame
if dependent_var in df_filtered_occupancy.columns and all(col in df_filtered_occupancy.columns for col in independent_vars):

    # Define the dependent variable
    y = df_filtered_occupancy[dependent_var]

    # Define the independent variables
    X = df_filtered_occupancy[independent_vars]

    # Handle missing values by dropping rows with NaNs in either X or y
    # Concatenation ensures X and y remain properly aligned
    combined_data = pd.concat([X, y], axis=1).dropna()

    # Separate independent variables after removing missing values
    X = combined_data[independent_vars]

    # Separate dependent variable after removing missing values
    y = combined_data[dependent_var]

    # Add an intercept term
    X = sm.add_constant(X)

    # Create the OLS regression model
    model = sm.OLS(y, X)

    # Fit the model and estimate regression coefficients and statistics
    results = model.fit()

    # Print the regression summary, including coefficients and p-values
    print("OLS Regression Results (P-values):")
    print(results.summary())

else:
    # Print an error message if required columns are missing
    print(f"One or more required columns ({dependent_var} or {independent_vars}) not found in the DataFrame.")

# This model represents the initial multivariate regression model.
# An adjusted R-squared of approximately 42% indicates a reasonable starting point.
# All included variables are statistically significant.
# Categorical variables and location-related measures have not yet been included,
# and may further improve the explanatory power of the model.
```

```python
# Create distance-based location variables by computing distances from listings to key landmarks in Munich
import numpy as np

# Compute great-circle distances between two geographic points using the Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in kilometers

    # Convert latitude and longitude to radians
    lat1_rad = np.radians(lat1)
    lon1_rad = np.radians(lon1)
    lat2_rad = np.radians(lat2)
    lon2_rad = np.radians(lon2)

    # Compute differences in coordinates
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad

    # Apply the Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    # Return distance in kilometers
    return R * c


# Define latitude and longitude of key landmarks in Munich
theresienwiese_lat, theresienwiese_lon = 48.13198, 11.54935   # Oktoberfest venue
central_station_lat, central_station_lon = 48.14023, 11.55836  # Central station
marienplatz_lat, marienplatz_lon = 48.13743, 11.57549         # City center (Marienplatz)

# Calculate distance from each listing to the Oktoberfest venue
df_filtered_occupancy['distance_to_oktoberfest'] = haversine(
    df_filtered_occupancy['latitude'],
    df_filtered_occupancy['longitude'],
    theresienwiese_lat,
    theresienwiese_lon
)

# Calculate distance from each listing to the central station
df_filtered_occupancy['distance_to_central_station'] = haversine(
    df_filtered_occupancy['latitude'],
    df_filtered_occupancy['longitude'],
    central_station_lat,
    central_station_lon
)

# Calculate distance from each listing to the city center (Marienplatz)
df_filtered_occupancy['distance_to_city_center'] = haversine(
    df_filtered_occupancy['latitude'],
    df_filtered_occupancy['longitude'],
    marienplatz_lat,
    marienplatz_lon
)


# Verify that the new distance variables have been added correctly
print("New distance variables added to df_filtered_occupancy.")
print(
    df_filtered_occupancy[
        ['distance_to_oktoberfest', 'distance_to_central_station', 'distance_to_city_center']
    ].head()
)

# Location information is incorporated into the dataset by converting geographic coordinates
# into numerical distance variables, allowing spatial effects on estimated revenue
# to be analyzed within a regression framework.
```

```python
# Examine the relationship between listing price and distance to key landmarks
# using simple linear regression

from scipy.stats import linregress
import matplotlib.pyplot as plt
import seaborn as sns

# List of distance variables to analyze
distance_vars = [
    'distance_to_oktoberfest',
    'distance_to_central_station',
    'distance_to_city_center'
]

# Iterate over each distance variable
for dist_col in distance_vars:

    # Remove rows with missing values to ensure valid input for linregress
    temp_df_for_r2 = df_filtered_occupancy[
        [dist_col, 'price_cleaned']
    ].dropna()

    # Perform a simple linear regression between distance and price
    # slope, intercept, r_value, and p_value are returned
    if temp_df_for_r2.empty or len(temp_df_for_r2) <= 1:
        print(f"Not enough data to run regression for {dist_col}.")
        continue

    slope, intercept, r_value, p_value, std_err = linregress(
        temp_df_for_r2[dist_col],
        temp_df_for_r2['price_cleaned']
    )

    # Compute R-squared from the correlation coefficient
    r_squared = r_value**2

    # Plot scatter points and the fitted regression line
    plt.figure(figsize=(10, 6))

    sns.regplot(
        data=df_filtered_occupancy,
        x=dist_col,
        y='price_cleaned',
        scatter_kws={'alpha': 0.4, 's': 20},
        line_kws={'color': 'red'}
    )

    plt.title(f'{dist_col} vs. Price (cleaned)')
    plt.xlabel(f'{dist_col} (km)')
    plt.ylabel('Price (cleaned)')

    plt.grid(True)

    # Display R-squared value on the plot
    plt.text(
        0.05, 0.95,
        f'R-squared = {r_squared:.2f}',
        transform=plt.gca().transAxes,
        fontsize=12,
        verticalalignment='top'
    )

    plt.show()

# The plot suggests that listings closer to the respective landmark tend to have higher prices.
# Although the R-squared value is relatively low, the variable still shows explanatory power
# and may be included in the final model for predicting estimated revenue.
```

```python
# Create a numerical feature representing the number of amenities for each listing
import ast # Library for safely evaluating string representations of Python literals

# Define a function to count the number of amenities for each listing
def count_amenities(amenities_str):
    # Return 0 if the amenities value is missing
    if pd.isna(amenities_str):
        return 0
    try:
        # Safely convert the string representation of a list into a Python list
        amenities_list = ast.literal_eval(amenities_str)
        # Return the number of amenities
        return len(amenities_list)
    except (ValueError, SyntaxError):  # Handle malformed or invalid list strings
        return 0

# Apply the function to create a new numerical variable representing the number of amenities for each listing
df_filtered_occupancy['num_amenities'] = (
    df_filtered_occupancy['amenities'].apply(count_amenities)
)

# Display a preview to verify the new variable
print(df_filtered_occupancy[['amenities', 'num_amenities']].head())
```

# Step 3. Model Validation

```python
# Prepare the dataset for Model 1 (to be used in k-fold cross-validation)
import statsmodels.api as sm

# Define the dependent variable
dependent_var_model1 = 'estimated_revenue_l365d'
y_model1 = df_filtered_occupancy[dependent_var_model1]

# Specify the independent variables for Model 1
X_model1_cols = [
    'beds', 'number_of_reviews_ltm', 'accommodates', 'bedrooms',
    'number_of_reviews_ly', 'number_of_reviews_l30d', 'bathrooms',
    'maximum_nights', 'availability_365', 'maximum_maximum_nights',
    'minimum_maximum_nights', 'review_scores_cleanliness',
    'maximum_nights_avg_ntm', 'availability_30'
]

# Create the independent-variable matrix
X_model1 = df_filtered_occupancy[X_model1_cols]

# Concatenate X and y and drop rows with missing values (ensures X and y are aligned for modeling and validation)
combined_model1_data = pd.concat([X_model1, y_model1], axis=1).dropna()

# Split cleaned data back into X and y
X_model1_cleaned = combined_model1_data[X_model1_cols]
y_model1_cleaned = combined_model1_data[dependent_var_model1]

#Add an intercept term
X_model1_cleaned = sm.add_constant(X_model1_cleaned)

# Confirm the prepared data shapes
print("Data for Model 1 prepared successfully.")
print(f"Shape of X_model1_cleaned: {X_model1_cleaned.shape}")
print(f"Shape of y_model1_cleaned: {y_model1_cleaned.shape}")
```

```python
# Evaluate the predictive performance of Model 1 (numerical variables only) using k-fold cross-validation
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_squared_error

n_splits = 5  # Number of folds
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize lists to store performance metrics for each fold
model1_r2_scores = []
model1_mse_scores = []

print(f"Performing {n_splits}-fold cross-validation for Model 1...")
print("--------------------------------------------------")

# Iterate over each fold
for fold, (train_index, test_index) in enumerate(kf.split(X_model1_cleaned)):
    # Split data into training and testing sets
    X_train, X_test = X_model1_cleaned.iloc[train_index], X_model1_cleaned.iloc[test_index]
    y_train, y_test = y_model1_cleaned.iloc[train_index], y_model1_cleaned.iloc[test_index]

    # Fit an OLS regression model on the training data
    model = sm.OLS(y_train, X_train)
    results = model.fit()

    # Generate predictions on the test data
    predictions = results.predict(X_test)

    # Compute performance metrics for the current fold
    r2 = r2_score(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)

    # Store the Mean Squared Error
    model1_mse_scores.append(mse)

    # Print fold-level results
    print(f"Fold {fold + 1}:")
    print(f"  Mean Squared Error: {mse:.2f}")

print("\n--------------------------------------------------")
print("Cross-validation for Model 1 completed.")

# Report the average Mean Squared Error across all folds
print(f"Average Mean Squared Error for Model 1: {np.mean(model1_mse_scores):.2f}")
```

```python
# Prepares the input and output data for an extended regression model (Model 2) by incorporating encoded categorical variables.

# Copy the DataFrame to avoid chained assignment issues
df_model_2 = df_filtered_occupancy.copy()

# Convert boolean-like categorical variables ('t'/'f') to 1/0
# Missing values are set to 0 for consistency
for col in ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified']:
    df_model_2.loc[:, col] = df_model_2[col].map({'t': 1, 'f': 0}).fillna(0).astype(int)

# One-hot encode selected categorical variables
# Fill missing categories with 'Unknown' before encoding
for col in ['neighbourhood_cleansed', 'property_type', 'room_type']:
    df_model_2.loc[:, col] = df_model_2[col].fillna('Unknown')

# Create dummy variables (drop_first=True reduces dummy-variable multicollinearity)
df_encoded_cats_model2 = pd.get_dummies(
    df_model_2[['neighbourhood_cleansed', 'property_type', 'room_type']],
    drop_first=True,
    prefix=['neighbourhood', 'property_type', 'room_type']
)

# Define the dependent variable for Model 2
dependent_var_model2 = 'estimated_revenue_l365d'

# Baseline numerical variables (from Model 1)
numerical_independent_vars_model2 = [
    'beds', 'number_of_reviews_ltm', 'accommodates', 'bedrooms',
    'number_of_reviews_ly', 'number_of_reviews_l30d', 'bathrooms',
    'maximum_nights', 'availability_365', 'maximum_maximum_nights',
    'minimum_maximum_nights', 'review_scores_cleanliness',
    'maximum_nights_avg_ntm', 'availability_30'
]

# Add location and amenities features
numerical_independent_vars_model2.extend([
    'distance_to_oktoberfest', 'distance_to_central_station', 'distance_to_city_center',
    'num_amenities'
])

# Combine numerical variables with boolean host features
X_model2_cols = numerical_independent_vars_model2 + [
    'host_is_superhost', 'host_has_profile_pic', 'host_identity_verified'
]

# Build X by concatenating selected columns with one-hot encoded categorical variables
X_model2 = pd.concat([df_model_2[X_model2_cols], df_encoded_cats_model2], axis=1)
y_model2 = df_model_2[dependent_var_model2]

# Drop rows with missing values to keep X and y aligned
combined_model2_data = pd.concat([X_model2, y_model2], axis=1).dropna()

X_model2_cleaned = combined_model2_data[X_model2.columns]
y_model2_cleaned = combined_model2_data[dependent_var_model2]

# Ensure all predictors are numeric for statsmodels
X_model2_cleaned = X_model2_cleaned.astype(float)

# Add an intercept term
X_model2_cleaned = sm.add_constant(X_model2_cleaned)

# Confirm prepared data shapes
print("Data for Model 2 prepared successfully.")
print(f"Shape of X_model2_cleaned: {X_model2_cleaned.shape}")
print(f"Shape of y_model2_cleaned: {y_model2_cleaned.shape}")
```

```python
# Fit an extended OLS regression model incorporating numerical, categorical, and engineered location/amenities features

# Create a working copy to avoid chained assignment warnings
df_model = df_filtered_occupancy.copy()

# Encode boolean-like host attributes ('t'/'f') as 1/0; treat missing values as 0
for col in ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified']:
    df_model.loc[:, col] = df_model[col].map({'t': 1, 'f': 0}).fillna(0).astype(int)

# Prepare categorical variables for one-hot encoding by imputing missing categories as 'Unknown'
for col in ['neighbourhood_cleansed', 'property_type', 'room_type']:
    df_model.loc[:, col] = df_model[col].fillna('Unknown')

# One-hot encode categorical variables; drop the first level to reduce dummy-variable multicollinearity
df_encoded_cats = pd.get_dummies(
    df_model[['neighbourhood_cleansed', 'property_type', 'room_type']],
    drop_first=True,
    prefix=['neighbourhood', 'property_type', 'room_type']
)

# Specify dependent variable and baseline numerical predictors (from Model 1)
dependent_var = 'estimated_revenue_l365d'
numerical_independent_vars_base = [
    'beds', 'number_of_reviews_ltm', 'accommodates', 'bedrooms',
    'number_of_reviews_ly', 'number_of_reviews_l30d', 'bathrooms',
    'maximum_nights', 'availability_365', 'maximum_maximum_nights',
    'minimum_maximum_nights', 'review_scores_cleanliness',
    'maximum_nights_avg_ntm', 'availability_30'
]

# Extend predictors with engineered location and amenities features
numerical_independent_vars_base.extend([
    'distance_to_oktoberfest', 'distance_to_central_station',
    'distance_to_city_center', 'num_amenities'
])

# Combine numerical predictors with binary host attributes
X_selected_cols = numerical_independent_vars_base + [
    'host_is_superhost', 'host_has_profile_pic', 'host_identity_verified'
]

# Construct design matrix X and response vector y (including encoded categorical variables)
X = pd.concat([df_model[X_selected_cols], df_encoded_cats], axis=1)
y = df_model[dependent_var]

# Drop rows with missing values to ensure X and y remain aligned
combined_data_cleaned = pd.concat([X, y], axis=1).dropna()
X_final = combined_data_cleaned[X.columns]
y_final = combined_data_cleaned[dependent_var]

# Ensure all predictors are numeric for statsmodels
X_final = X_final.astype(float)

# Fit OLS regression if sufficient observations remain after cleaning
if not X_final.empty and len(X_final) > 1:
    # Add intercept term
    X_final = sm.add_constant(X_final)

    # Fit the OLS model and report the regression summary
    model_new = sm.OLS(y_final, X_final)
    results_new = model_new.fit()

    print("OLS Regression Results with Additional Categorical and Engineered Features:")
    print(results_new.summary())
else:
    print("Insufficient data after removing missing values; OLS regression was not performed.")

# This specification includes baseline numerical predictors, engineered distance/amenities features,
# binary host attributes, and one-hot encoded categorical variables.
# Although the adjusted R-squared improves relative to Model 1, further refinement is warranted,
# as several predictors remain statistically insignificant (p > 0.05).
```

```python
# Refine the extended regression model using backward elimination to retain only statistically significant predictors

# X_final and y_final are already defined and cleaned in a previous step
# X_final already includes an intercept term ('const')

# Initialize the model with the full set of predictors
current_X = X_final.copy()
current_y = y_final.copy()

print(f"Starting with {len(current_X.columns) - 1} variables (excluding the intercept).")

# Perform backward elimination based on p-values
while True:
    # Fit OLS model with the current set of variables
    model = sm.OLS(current_y, current_X)
    results = model.fit()

    # Extract p-values for all predictors except the intercept
    p_values_for_selection = results.pvalues.drop('const', errors='ignore')

    # Stop if no predictors remain
    if p_values_for_selection.empty:
        break

    # Identify the predictor with the largest p-value
    max_p_value = p_values_for_selection.max()
    most_insignificant_var = p_values_for_selection.idxmax()

    # Remove the predictor if it exceeds the significance threshold
    if max_p_value > 0.05:
        print(f"Removing '{most_insignificant_var}' (p-value = {max_p_value:.3f})")
        current_X = current_X.drop(columns=[most_insignificant_var])

        # Stop if only the intercept remains
        if current_X.empty or (len(current_X.columns) == 1 and 'const' in current_X.columns):
            break
    else:
        # All remaining predictors are statistically significant
        break

# Use the final fitted model after backward elimination
results_refined = results

print(
    f"\nFinal model retains {len(results_refined.pvalues) - ('const' in results_refined.pvalues.index)} "
    "statistically significant predictors (excluding the intercept)."
)

print("Refitted OLS model with only statistically significant variables:")
print(results_refined.summary())

# Store refined predictors and response for downstream analysis
X_model_refined = current_X
y_model_refined = current_y

# This final specification maintains a high adjusted R-squared while ensuring that
# all retained predictors are statistically significant at the 5% level.
```

```python
# Generate predicted revenues from the refined OLS model and compute potential

# Generate predictions using the refined model
predicted_revenue = results_refined.predict(X_model_refined)

# Align predictions with the original DataFrame index
predicted_revenue_series = pd.Series(predicted_revenue, index=y_model_refined.index)

# Add predicted revenue to the dataset
df_with_predictions = df_filtered_occupancy.copy()
df_with_predictions['predicted_estimated_revenue_l365d'] = predicted_revenue_series

# Compute potential as the gap between predicted and actual estimated revenue
df_with_predictions['potential'] = (
    df_with_predictions['predicted_estimated_revenue_l365d']
    - df_with_predictions['estimated_revenue_l365d']
)

# Update the main DataFrame
df_filtered_occupancy = df_with_predictions

print("A new column 'potential' has been added to df_filtered_occupancy.")
print(
    df_filtered_occupancy[
        ['estimated_revenue_l365d', 'predicted_estimated_revenue_l365d', 'potential']
    ].head()
)

# The 'potential' metric identifies listings that underperform relative to
# their predicted revenue based on model characteristics.
```

```python
# Prepare and standardize features for K-means clustering to identify attractive Airbnb listings
from sklearn.preprocessing import StandardScaler


# Select variables used for clustering
df_kmeans_features = df_filtered_occupancy[
    ['estimated_revenue_l365d', 'potential']
].copy()

# Remove observations with missing values
df_kmeans_features.dropna(inplace=True)

# Initialize the standard scaler
scaler = StandardScaler()

# Standardize features to ensure equal scale for K-means clustering
df_scaled_features = pd.DataFrame(
    scaler.fit_transform(df_kmeans_features),
    columns=df_kmeans_features.columns,
    index=df_kmeans_features.index
)

print("Features for K-means clustering prepared and scaled successfully.")
print("First 5 rows of df_kmeans_features:")
print(df_kmeans_features.head())
print("\nFirst 5 rows of df_scaled_features:")
print(df_scaled_features.head())

# K-means clustering is applied to segment listings based on estimated revenue and underperformance potential
# in order to identify the most attractive investment opportunities.
```

```python
# Select an appropriate number of clusters (K) for K-means using inertia (elbow) and silhouette scores
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


# Store evaluation metrics across different K values
inertia_values = []
silhouette_scores = []

# Evaluate K from 1 to 10 (silhouette is defined only for K >= 2)
k_values = range(1, 11)

print("Calculating inertia and silhouette scores for different K values...")

# Fit K-means for each K and collect metrics
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(df_scaled_features)

    # Inertia: within-cluster sum of squares
    inertia_values.append(kmeans.inertia_)

    # Silhouette score: computed only when K > 1
    if k > 1:
        silhouette_avg = silhouette_score(df_scaled_features, kmeans.labels_)
        silhouette_scores.append(silhouette_avg)

print("Done calculating scores.")

# Plot inertia (elbow) and silhouette score across K values
plt.figure(figsize=(12, 5))

# Elbow curve (start from K=2 since K=1 is not informative for elbow selection)
plt.subplot(1, 2, 1)
plt.plot(k_values[1:], inertia_values[1:], marker='o')
plt.title('Elbow Method for Selecting K (Inertia)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(k_values[1:])
plt.grid(True)

# Silhouette scores (defined only for K >= 2)
plt.subplot(1, 2, 2)
plt.plot(k_values[1:], silhouette_scores, marker='o')
plt.title('Silhouette Scores for Selecting K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values[1:])
plt.grid(True)

plt.tight_layout()
plt.show()

# The elbow method suggests K ≈ 4, while the silhouette score peaks at K ≈ 2.
# We proceed with K = 3 as a practical compromise that yields interpretable segments
# aligned with the project objective.
```

```python
# Apply K-means clustering with the selected number of clusters and visualize cluster characteristics

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Select the number of clusters based on prior elbow and silhouette analysis
optimal_k = 3
print(f"Applying K-Means clustering with optimal_k = {optimal_k}...")

# Initialize and fit the K-means model on the scaled feature space
kmeans_model = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
kmeans_model.fit(df_scaled_features)

# Assign cluster labels to the original (unscaled) feature DataFrame
df_kmeans_features['cluster'] = kmeans_model.labels_
print("Cluster labels assigned successfully.")

# Visualize clusters in the estimated revenue–potential space
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=df_kmeans_features,
    x='estimated_revenue_l365d',
    y='potential',
    hue='cluster',
    palette='viridis',
    s=50,
    alpha=0.7
)

plt.title(f'K-Means Clusters (K = {optimal_k})')
plt.xlabel('Estimated Revenue (Last 365 Days)')
plt.ylabel('Potential (Predicted − Actual Revenue)')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()

# Summarize cluster characteristics using mean values
print("\nCluster-Level Summary Statistics (Mean Values):")
cluster_summary = (
    df_kmeans_features
    .groupby('cluster')[['estimated_revenue_l365d', 'potential']]
    .mean()
)
print(cluster_summary)

# Clusters can be interpreted in terms of attractiveness:
# high potential with relatively low revenue indicates more attractive listings,
# while low potential clusters are less attractive under the current criteria.
```
